## Missing Data Handling

In this notebook, some inbuilt method offered by pandas are illustrated to handle missing datas. 
<br>This notebook also uses the same dataset used in the previous notebook. However, you can download the data from [here](https://www.kaggle.com/c/titanic/data).

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
df.info()
```

```python
None == None
```

```python
np.nan == np.nan
```

```python
(df['Age'] == np.nan).head()
```

```python
df[df['Age'] == np.nan]
```

```python
df[pd.isnull(df['Age'])].head()
```

```python
df[pd.notnull(df['Age'])].head()
```

```python
df.fillna(555555).head()
```

```python
df.fillna('').head()
```

```python
df.fillna(df.mean()).head()
```

```python
df.fillna(method='pad', limit=1).head(10)
```
