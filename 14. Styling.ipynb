```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(10, 6), columns=list('ABCDEF'))
df.iloc[5,1] = np.nan
```

```python
df
```

```python
def color_negative_red(val):

    color = 'red' if val < 0 else 'blue'
    return 'color: %s' % color
```

Element-wise, thus **applymap**

```python
s = df.style.applymap(color_negative_red)
s
```

Column wise, thus **apply**

```python
def highlight_max(s):

    is_max = s == s.max()
    #print(is_max)
    return ['background-color: yellow' if v else '' for v in is_max]
```

```python
df.style.apply(highlight_max)
```

### Predefined Functions

##### Default axis [=0]

```python
df.style.highlight_max()
```

##### axis = 1

```python
df.style.highlight_max(axis=1)
```

##### Highlighting Null

```python
df.style.highlight_null()
```

##### Using both style

```python
df.style.applymap(color_negative_red).apply(highlight_max)
```

##### Finer control: Slicing

```python
df.style.apply(highlight_max, subset=['B', 'C', 'D'])
```

##### Finer Control: Display Values

```python
df.style.format("{:.4f}")
```

```python
df.style\
  .applymap(color_negative_red)\
  .apply(highlight_max)\
  .set_precision(2)
```

#### Colormap

```python
import seaborn as sns

cm = sns.light_palette("y", as_cmap=True)

s = df.style.background_gradient(cmap=cm)
s
```
