```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Lets visualise some random data

##### Line plot,  Bar plot,  Histogram,  Area plot,  Scatter Plot

#### Series

```python
ser = pd.Series(np.random.randn(1000))
```

```python
ser = ser.cumsum()
```

```python
ser.plot()
```

```python
ser.plot(figsize=(16,4))
```

#### Dataframe

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=list('ABCD'))
```

```python
df = df.cumsum()
```

```python
df.plot(figsize=(16,4))
```

#### Using Subplot

```python
df.plot(subplots=True,figsize=(16,10), colormap='gist_rainbow')
```

```python
df.plot(subplots=True, layout=(2, 2), figsize=(16, 8), colormap='viridis');
```

#### Bar plot

```python
df2 = pd.DataFrame(np.random.rand(10, 3), columns=['a', 'b', 'c'])
```

```python
df2.plot.bar(figsize=(16,5));
```

#### Stacked bar plot

```python
df2.plot.bar(figsize=(16,5),stacked=True);
```

##### Histogram

```python
df.plot.hist(alpha=0.75, figsize=(6,5))
```

##### Box Plot

```python
df.plot.box()
```

##### Area Plot
Auto Stacked = True

```python
df.abs().plot.area(figsize=(6,5))
```

```python
df.plot.area(stacked=False)
```

##### Scatter plot

```python
df.plot.scatter(x='D', y='B', figsize=(8,8))
```

```python
df.iloc[:200].plot.scatter(x='D', y='B', figsize=(8,8),c='purple', s=50)
```

```python
df.iloc[:200].plot.scatter(x='D', y='B', figsize=(8,8),c='purple', s=df['C']*10)
```
