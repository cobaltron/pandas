```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(37)
data = np.random.randn(150).cumsum()
```

```python
price = pd.Series(data,index=pd.date_range('2016-1-1', periods=150, freq='D'))
```

```python
plt.figure(figsize=(16,4))
plt.plot(price.index, price, 'k')
```

```python
ma = price.rolling(20).mean()
```

```python
plt.figure(figsize=(16,4))
plt.plot(price.index, price, 'k')
plt.plot(ma)
```

```python
ma = price.rolling(20).mean().shift(-10)
```

```python
plt.figure(figsize=(16,4))
plt.plot(price.index, price, 'k')
plt.plot(ma)
```

```python
ma = price.rolling(20).mean().shift(-10)
plt.figure(figsize=(16,4))
plt.plot(price.index, price, 'k')
plt.plot(ma)
```

```python
price.rolling(10).std().head()
```

```python
ma = price.rolling(20).mean().shift(-10)
mstd = price.rolling(20).std().shift(-10)
```

```python
plt.figure(figsize=(16,5))
plt.grid()
plt.plot(price.index, price, 'k', label='raw')
plt.plot(ma.index, ma, 'b', label='rolling_mean')
plt.fill_between(mstd.index, ma-1.5*mstd, ma+1.5*mstd, color='grey', alpha=0.25)
plt.title('Implementation of Rolling statistics')
plt.legend(fontsize=15)
```

```python
plt.figure(figsize=(16,5))
plt.grid()
plt.plot(price.rolling(10).sum())
```

```python
plt.figure(figsize=(16,5))
plt.grid()
plt.plot(price.rolling(10).median())
```

```python
plt.figure(figsize=(16,5))
plt.grid()
plt.plot(price.rolling(10).corr())
```
