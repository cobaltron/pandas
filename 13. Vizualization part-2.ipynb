```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Hexagonal Bin Plot

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.head()
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25, figsize=(8,6))
```

#### Pie Chart

###### Using Series

```python
ser = pd.Series(np.random.rand(6), index=list('abcdef'))
```

```python
ser.plot.pie(figsize=(8, 8))
```

```python
ser.plot.pie(figsize=(8, 8), autopct='%.3f', fontsize=15)
```

```python
ser.plot.pie(labels=['Sunday','Monday','Tuesday','Wednessday','Thursday','Friday'],figsize=(8, 8), autopct='%.3f', fontsize=15)
```

###### Using Dataframe

```python
df.iloc[:4].abs().plot.pie(subplots=True, figsize=(12, 6),colormap='viridis')
```

### Plotting tools

```python
from pandas.plotting import scatter_matrix
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=['a', 'b', 'c', 'd'])
scatter_matrix(df, alpha=0.2, figsize=(12,12), diagonal='kde')
```

#### Density Plots

```python
ser = df['a']
```

```python
ser.plot.kde()
```

```python
ser.mean()
```

#### Andrews Curves

```python
from pandas.plotting import andrews_curves
```

```python
df['label'] = np.random.randint(0,3,1000)
```

```python
plt.figure(figsize=(16,6))
andrews_curves(df[:20], 'label',colormap='viridis')
```

```python
from pandas.plotting import radviz
```

```python
plt.figure(figsize=(10,10))
radviz(df[:200], 'label',colormap='winter')
```
