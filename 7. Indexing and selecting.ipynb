## Pandas Series & Dataframe Indexing and Selection

**`.loc`** is primarily label based
<br>.**`iloc`** is primarily integer position based (from 0 to length-1 of the axis)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/data.csv')
df.head()
```

```python
df.iloc[50]
```

```python
df.iloc[[1,2]]
```

```python
df.loc[50]
```

```python
df['SYMBOL'].head(10)
```
